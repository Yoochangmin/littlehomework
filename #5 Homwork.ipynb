# **[5번 문제]**  
4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.  
- 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.  
- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.  
>>> u = s | t  
>>> print(u)  
MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}  
- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.  
>>> u = s & t  
>>> print(u)  
MySet: {2, 3}  
- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.  
>>> s = MySet([1, 2, 3])  
>>> t = MySet([3, 4, 5])  
>>> u = s - t  
>>> print(u)  
MySet: {1, 2}

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
 
    def eliminate_duplicate(l): #중복되는 원소 방지
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    def __or__(self, other): #합집합
        temp = []
        for i in self:
            temp.append(i)        
        for j in other:
            temp.append(j)
        MySet.eliminate_duplicate(temp)
        return MySet(temp)

    def __and__(self, other): #교집합
        temp = []
        for i in self:
            if i not in  temp and i in other:
                temp.append(i)
        return MySet(temp)
      
    def __sub__(self, other): #차집합(s-t)
        temp = []
        for i in self:
            if i not in temp and i not in other:
                temp.append(i)
        return MySet(temp)

    
if __name__ == "__main__":
    s = MySet([1, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    u = s | t
    print (u)
    l = s & t
    print(l)
    k = s - t
    print(k)
```

**Answer**  
합집합을 얻기 위해 temp라는 빈 리스트를 설정하였다. temp에는 self와 other의 원소를 append를 통해 넣었고 미리 설정해둔 eliminate_duplicate를 통해 중복된 원소는 제거하여 출력되도록 하였다. 교집합은 self를 반복문을 통해 돌며 temp에는 없고 other에는 있는 원소들을 temp에 넣도록 하였다. 마지막으로 차집합은 반복문을 통해 self를 돌며 temp에 없고, other에도 없는 원소들을 넣도록 하였다.
