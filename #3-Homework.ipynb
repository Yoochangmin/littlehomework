    s = '  first star   :   second star   :    third star  '
    l = ['first star', 'second star', 'third star']
* 1) for ~ in 문을 이용한 코드를 작성하시오.
* 2) 리스트 내포(list comprehension)를 이용한 코드를 작성하시오.

```python
#1 방법
s = '  first star   :   second star   :    third star  '
s = s.split(":")
l = []
for i in s:
    i = i.strip()
    l.append(i)
print(l)
```

```python
#2 방법
s = '  first star   :   second star   :    third star  '
s = s.split(":")
l = [i.strip() for i in s]
print(l)
```

# #5다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.
* 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
 *  '>>>'addall([1])
 * 1
 * '>>>' addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
 * 45
* 2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
 * '>>>' addallodd([1])
 * 1
 * '>>>' addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
 * 25
* [참고 1]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)
* [참고 2]: sum() 등의 내장함수 사용 가능함

```python
def addall(list):
    total=0
    for i in list:
        total+= i
    return print(total)
def addallodd(list):
    total=0
    for i in list:
        if i % 2 != 0:
            total+= i
    return print(total)
addall([1,2,3,4,5,6,7,8,9])
addallodd([1,2,3,4,5,6,7,8,9])
```

```python
def addall_(list): 
    return print(sum([i for i in list]))
def addallodd_(list):
    return print(sum([i for i in list if i%2 !=0]))

addall_([1,2,3,4,5,6,7,8,9])
addallodd_([1,2,3,4,5,6,7,8,9])

```

* 1. for문을 이용할 때
 * 함수내부에 값을 리스트 값을 모두 더할 total 변수를 선언한 뒤 for 문을 이용하여 list의 인덱스 값들을 차례대로 불러온뒤 total에 i값을 계속 해서 더하는 식으로 addall()함수를 만들고 addllodd()함수는 for문 뒤에 if i %2 != 0이라는 조건문을 추가하여 0으로 나누어 떨어지지 않으면 total에 계속해서 i값을 더하게 구성

* 2. 리스트내포 이용
 * return 값으로 print(sum())을 이용하고 sum내포 인자로 리스트 내포로 i for i in list를 통해 리스트안에 있는 인덱스 값들을 차레대로 i에 추가 한뒤 모든 요소를 더함
 * addallodd 함수는 리스트 내포로 i for i in list에 추가로 조건문 if i%2 != 0을 추가하여 홀수만 삽입할 수 있도록 구성

* [참고]: d.keys(), d.values()를 통해 리스트를 얻어낸 후 리스트가 지니고 있는 sort(cmp=None, key=None, reverse=False)함수를 활용하시오.
* 1) 키의 알파벳 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.
* 2) 키의 알파벳 내림차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.
* 3) 값의 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.

```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
key = list(d.keys())
value = list(d.values())
#1 키의 알파벳 오름차순 순서
KeyUp= [item for item in zip(key, value)]
KeyUp.sort(key = lambda x:x[0])
print(KeyUp)

#2 키의 알파벳 내림차순 순서
KeyDown= [item for item in zip(key, value)]
KeyDown.sort(key = lambda x:x[0], reverse =True)
print(KeyDown)

#3 값의 오름차순 순서
ValueUp= [item for item in zip(key, value)]
ValueUp.sort(key = lambda x:x[1])
print(ValueUp)
#4 값의 내림차순 순서
ValueDown= [item for item in zip(key, value)]
ValueDown.sort(key = lambda x:x[1], reverse =True)
print(ValueDown)
```

* 이 문제에서 숫자 하나가 주어질 때, 이 숫자가 Strange인지 판단하는 함수 isStrange(num)을 작성하시오.
* isStrange(num) 함수가 재귀적인 함수가 되도록 작성해 보세요.


* def isStrange(num):
    '''
    함수 포맷 변경하지 마세요.
    재귀적인 함수는 함수 안에서 자기 자신 함수를 호출하는 것을 말합니다.
    '''
   return True # or False or isStrange(매개변수 숫자)
[예시]
1. num = 0                      =>    True
2. num = 56                     =>    True
3. num = 100                    =>    False
4. num = 7882602506492313600    =>    True
5. num = 6012302549522098651    =>    False

```python
def length(X):
    digit = []
    for i in str(X):
        digit.append(i)
    return len(digit)


def isStrange(num):
    if num < 0:
        return False
    else:
        if length(num) == 1:
            return True
        elif length(num) > 1:
            if num % length(num) == 0: 
                if isStrange(num//length(num)):
                    return True
                else:
                    return False
            else:
                return False
print(isStrange(56))
```

```python
import time
def get_divisor(n): #약수를 만드는 함수
    n = int(sum(n))
    divisors = []
    divisors_back = [] 

    for i in range(1, int(n**(1/2)) + 1): 
        if (n % i == 0):            
            divisors.append(i)
            if (i != (n // i)): 
                divisors_back.append(n//i)

    return divisors + divisors_back[::-1]

def solve(a):
    div = get_divisor(a)

    result = []
    x = 0
    for i in div:
        check = True
        for j in a:     
            if j > i:    #a 인덱스 값이 약수보다 크다면
                check = False
                break    #약수의 다음 값으로 이동
                
            else:  # 작거나 같다면
                x += j
                if x == i:
                    x = 0
                    
                elif x > i:
                    x = 0
                    check = False
                    break
        if check:  
            result.append(i)
    return result
    
        
print(solve([1,1,1,1,1,1,1,1]))
print(solve([1,2,1,1,1,2,1,3]))
print(solve([1,2,3,5,4,1,2]))
start = time.time()
print(solve([9999]*10**5))
end = time.time()
print("함수 실행시간 : {}".format(end-start) )
```

* 약수를 만드는 함수인 gey_divisor함수를 만들기 위해 임의의 자연수 N의 약수들 중에서 두 약수의 곱이 N이 되는 약수 A와 약수 B는 반드시 존재한다'는 규칙이 존재하기 때문에, 자연수 N의 제곱근까지의 약수를 구하면 그 짝이 되는 약수는 자동으로 구해진다. (i, n//i)는 정리를 이용 n은 매개변수로 받은 인자 a리스트 요소들의 총합
* 반복문을 통해 1부터 자연수 n의 제곱근 까지 n이 i로 나누어 떨어지면 i를 리스트 divisors에 삽입하고 나누어 떨어지면서 i가 n을 i로나눈 정수가 아니면 divisors_back에 삽입
* return값으로는 divisors 리스트에 divisors_back을 거꿀로 출력한 리스트를 삽입

3. 이중 반복문을 통해 약수의 요소들을 i에 차례대로 대입시키고 a리스트의 요소들 값 즉 그룹별 사람들의 수를 j에 대입시키며 비교 시작
4. check는 a의 요소들 값이 모두 j에 반환 됫을 때 result에 i값을 대입 시키기 위해 설정
5. j(그룹의 사람수)가 i(버스의 정원 수)보다 크다면 check를 False로 반환시킨후 반복문을 탈출하여 다음 약수값으로 비교 시작
6. j(그룹의 사람수)가 i(버스의 정원 수)보다 작거나 같다면 x에 j(그룹의 사람수)값를 중첩시킨 뒤 만약 x(버스에 탄 사람수) == i(버스의 정원 수)라면 x를 다시 0으로 초기화 x(버스에 탄사람수)가 i(버스의 정원 수)보다 크다면 x를 0으로 다음 인덱스 값을 비교하기 위해 초기화 시킨뒤 check를 False로 반환 시킨 수 i의 다음 요소 값을 비교하기 위해 반복문 탈출
7. j가 a리스트의 마지막 인덱스 까지 check가 True라면 그 때의 i값을 result에 삽입

```python
import urllib.request
import re

import urllib.request

path = 'http://www.daum.net'

with urllib.request.urlopen(path) as response:
	html = response.read().decode("utf-8")

body = re.search('<body.*</body>', html, re.I|re.S)

body = body.group()
body = re.sub('<a href.*?>.*?">', '', body, 0, re.I|re.S)
body = re.sub('<script.*?>.*?</script>', '', body, 0, re.I|re.S)
body = re.sub('<style.*?>.*?</style>', '', body, 0, re.I|re.S)
body = re.sub('<.*?>', '', body, 0, re.I|re.S)

body = re.sub('[->/@+&#·:(){}→.;,~=?+]', '', body, 0, re.I|re.S)
body = re.sub('&nbsp;|\t|\r|\\|', '', body)
body = body.replace("[","")
body = body.replace("]","")

splitstring = body.split()
new_list = {}
for i in splitstring:
	try:
		new_list[i] += 1
	except:
		new_list[i] = 1
print (new_list)
print ('html = ', len(new_list))

```
